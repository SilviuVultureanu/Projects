```python
import pandas as pd

orders = pd.read_csv('shoefly (2).csv')



orders['shoe_source'] = orders.shoe_material.apply(lambda x: 'animal' if x == 'leather' else 'vegan')
```

```python
orders
```

```python
greeting = lambda row: 'Dear Mr. ' + row['last_name'] if row['gender'] == 'male' else 'Dear Ms. ' + row['last_name']
orders['greetings'] = orders.apply(greeting, axis=1)
orders
```
