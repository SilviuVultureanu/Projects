```python
import numpy as np
from matplotlib import pyplot as plt

# load in data
in_bloom = np.loadtxt(open("in-bloom.csv"), delimiter=",")
flights = np.loadtxt(open("flights.csv"), delimiter=",")

# Plot the histograms
plt.figure(1)
plt.subplot(211)
plt.hist(flights, range=(0, 365), bins=365, edgecolor='black')
#plt.hist(flights, range=(0, 365), bins=365)
plt.title('Flights per week')
plt.xlabel('# week')
plt.ylabel('# flights')

plt.subplot(212)
plt.hist(in_bloom, range=(0, 365), bins=365, edgecolor='black')
plt.title('Flowers in bloom per week')
plt.xlabel('# week')
plt.ylabel('# blooming flowers')

plt.tight_layout()
plt.show()
```
